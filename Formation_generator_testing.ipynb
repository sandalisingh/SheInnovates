```python
import numpy as np
import matplotlib.pyplot as plt
import Configurations as CF

from FormationGenerator import FormationGenerator
from InteractiveVoronoiPitch import InteractiveVoronoiPitch


def visualize_single_team(structure=CF.Formation.F_4_2_4,
                          shape=CF.Shape.NA,
                          mode=CF.Mode.ATTACKING,
                          team_side="left"):

    generator = FormationGenerator()

    # Build formation name
    formation_name = structure
    if shape != CF.Shape.NA:
        formation_name = f"{structure} {shape}"
    if mode != CF.Mode.UNKNOWN or mode != "":
        formation_name = f"{formation_name} {mode}"


    # Generate real team template
    team_coords = generator.generate_template_from_formation(
        formation_name=formation_name,
        team_side=team_side
    )

    team_coords = np.array(team_coords, dtype=float)

    # ---------------------------------------------------
    # Create dummy opponent (invisible minimal cluster)
    # ---------------------------------------------------
    # Put them far compressed so they don't affect view much
    dummy_coords = np.array([
        [104, 5],
        [104, 10],
        [104, 15],
        [104, 20],
        [104, 25],
        [104, 30],
        [104, 35],
        [104, 40],
        [104, 45],
        [104, 50],
        [104, 55],
    ], dtype=float)

    # If user wants team on right side
    if team_side == "right":
        home_coords = dummy_coords
        away_coords = team_coords
        home_form = "Dummy"
        away_form = formation_name
        home_mode = "none"
        away_mode = mode
    else:
        home_coords = team_coords
        away_coords = dummy_coords
        home_form = formation_name
        away_form = "Dummy"
        home_mode = mode
        away_mode = "none"

    # Launch interactive board
    InteractiveVoronoiPitch(
        home_coords,
        away_coords,
        home_form,
        away_form,
        home_mode,
        away_mode
    )

    plt.show()
```
